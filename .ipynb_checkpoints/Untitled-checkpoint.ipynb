```python
import xml.etree.ElementTree as ET

# Load the XML data from the file
file_path = 'santandercycle.xml'

try:
    tree = ET.parse(file_path)
    root = tree.getroot()
    
    # Print the root tag to confirm successful parsing
    print(f"Root tag: {root.tag}")
    
    # Extract data from XML
    stations = {}
    for station in root.findall('station'):
        station_id = station.find('id').text
        lat = float(station.find('lat').text)
        lon = float(station.find('long').text)
        stations[station_id] = (lat, lon)
    
    # Print extracted data for verification
    for station_id, coords in stations.items():
        print(f"Station ID: {station_id}, Latitude: {coords[0]}, Longitude: {coords[1]}")
        
except ET.ParseError as e:
    print(f"XML Parse Error: {e}")
except FileNotFoundError as e:
    print(f"File Not Found Error: {e}")
def print_xml_snippet(file_path, line_number, snippet_size=20):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        start = max(line_number - snippet_size // 2, 0)
        end = min(line_number + snippet_size // 2, len(lines))
        snippet = lines[start:end]
        for i, line in enumerate(snippet, start=start + 1):
            print(f"Line {i}: {line.strip()}")

# Example usage
print_xml_snippet('santandercycle.xml', 1449)
```

```python
import pandas as pd
import xml.etree.ElementTree as ET
from geopy.distance import great_circle

# Load the XML file and parse it
tree = ET.parse('santandercycle.xml')
root = tree.getroot()

# Extract station data from XML
station_data = {}
for station in root.findall('station'):
    station_id = station.find('id').text
    name = station.find('name').text
    lat = float(station.find('lat').text)
    lon = float(station.find('long').text)
    station_data[name] = {'lat': lat, 'lon': lon}

# Load the CSV file into a DataFrame
df = pd.read_csv('cycle02Jan2023-08Jan2023.csv')

# Function to map station name to latitude and longitude
def get_coordinates(station_name, data):
    if station_name in data:
        return data[station_name]['lat'], data[station_name]['lon']
    return None, None

# Add latitude and longitude columns based on station names
df['Start Latitude'], df['Start Longitude'] = zip(*df['Start station'].apply(lambda x: get_coordinates(x, station_data)))
df['End Latitude'], df['End Longitude'] = zip(*df['End station'].apply(lambda x: get_coordinates(x, station_data)))

# Check for missing or invalid coordinates and filter them out
df_cleaned = df.dropna(subset=['Start Latitude', 'Start Longitude', 'End Latitude', 'End Longitude'])

# Function to calculate distance
def calculate_distance(row):
    start_coords = (row['Start Latitude'], row['Start Longitude'])
    end_coords = (row['End Latitude'], row['End Longitude'])
    return great_circle(start_coords, end_coords).kilometers

# Add distance column to the cleaned DataFrame
df_cleaned['Distance (km)'] = df_cleaned.apply(calculate_distance, axis=1)

# Save the updated DataFrame to a new CSV file
df_cleaned.to_csv('cycle_with_distances_cleaned.csv', index=False)

print("Distance calculation completed and saved to cycle_with_distances_cleaned.csv")
```

```python
import pandas as pd
import sqlite3

# Step 1: Load the Excel and CSV files properly
bus_routes = pd.read_excel('C://Users//smk10/Desktop//dissertation//data//bus_routes.xlsx')  # Load Excel file
cleaned_bus_data = pd.read_csv('cleaned_merged_bus_data.csv')  # Load CSV file
all_buses_data = pd.read_excel('all_merged_buses_data.xlsx')  # Load Excel file

# Print column names to debug
print("Columns in all_buses_data:", all_buses_data.columns)
print("Columns in cleaned_bus_data:", cleaned_bus_data.columns)
print("Columns in bus_routes:", bus_routes.columns)

# Check for duplicate registration numbers
duplicates = all_buses_data[all_buses_data.duplicated(['registration number'], keep=False)]
print("Duplicate entries found:")
print(duplicates)

# Step 2: Filter for buses in bus_routes
bus_routes_filtered = bus_routes[bus_routes['Mode'] == 'bus']

# Ensure that the 'Routes ' column exists in all_buses_data and 'Route' in cleaned_bus_data
if 'Routes ' in all_buses_data.columns and 'Route' in cleaned_bus_data.columns:
    # Step 3: Merge all_buses_data with cleaned_bus_data on the "Routes " field
    merged_data = pd.merge(all_buses_data, cleaned_bus_data, left_on='Routes ', right_on='Route', how='left')
else:
    raise KeyError("Column 'Routes ' or 'Route' not found in the respective dataframes.")

# Ensure that the 'Line Name' column exists in both dataframes
if 'Line Name' in bus_routes_filtered.columns:
    # Step 4: Merge the result with bus_routes_filtered on "Line Name" to get route details
    final_merged_data = pd.merge(merged_data, bus_routes_filtered, left_on='Routes ', right_on='Line Name', how='left')
else:
    raise KeyError("Column 'Line Name' not found in bus_routes_filtered.")

# Ensure that the columns for Destination and Destination Code exist
if 'Destination' in final_merged_data.columns and 'Destination Code' in final_merged_data.columns and \
   'Stop_Name_x' in cleaned_bus_data.columns and 'Bus_Stop_Code_x' in cleaned_bus_data.columns:
    # Step 5: Link Destination and Destination Code with Stop Name and Bus_Stop_Code_x
    final_merged_data = pd.merge(final_merged_data,
                                 cleaned_bus_data,
                                 left_on=['Destination', 'Destination Code'],
                                 right_on=['Stop_Name_x', 'Bus_Stop_Code_x'],
                                 how='left', suffixes=('', '_stop'))
else:
    raise KeyError("Columns 'Destination', 'Destination Code', 'Stop_Name_x', or 'Bus_Stop_Code_x' not found.")

# Remove duplicate rows based on 'registration number'
final_merged_data = final_merged_data.drop_duplicates(subset=['registration number'])

# Step 6: Save to SQLite Database with "Registration Number" as the primary key
conn = sqlite3.connect('buses_database.db')

# Save to SQLite table
final_merged_data.to_sql('bus_data', conn, if_exists='replace', index=False)

# Create an index on Registration Number for faster querying
conn.execute('CREATE UNIQUE INDEX idx_registration_number ON bus_data ("registration number")')

# Close the connection
conn.close()

print("Database created successfully!")
```

```python
import pandas as pd

# Load the Excel and CSV files
all_merged_buses = pd.read_excel('all_merged_buses_data.xlsx')
cleaned_bus_data = pd.read_csv('cleaned_merged_bus_data.csv')
bus_routes = pd.read_excel('bus_routes.xlsx')

# Check for duplicate index values in all_merged_buses
print("Duplicate index values in all_merged_buses:", all_merged_buses.index[all_merged_buses.index.duplicated()])

# Remove duplicate index values if present
all_merged_buses = all_merged_buses[~all_merged_buses.index.duplicated(keep='first')]

# Repeat for cleaned_bus_data and bus_routes if necessary
print("Duplicate index values in cleaned_bus_data:", cleaned_bus_data.index[cleaned_bus_data.index.duplicated()])
cleaned_bus_data = cleaned_bus_data[~cleaned_bus_data.index.duplicated(keep='first')]

print("Duplicate index values in bus_routes:", bus_routes.index[bus_routes.index.duplicated()])
bus_routes = bus_routes[~bus_routes.index.duplicated(keep='first')]
# Expand the 'Routes ' column to handle multiple values
all_merged_buses_expanded = all_merged_buses.assign(
    Route=all_merged_buses['Routes '].str.split(',').explode()
).reset_index(drop=True)

# Ensure no duplicate columns after expansion
print("Columns in all_merged_buses_expanded:", all_merged_buses_expanded.columns)

# Merge expanded all_merged_buses with cleaned_bus_data on the route
merged_data = pd.merge(all_merged_buses_expanded, cleaned_bus_data, left_on='Route', right_on='Route', how='left')

# Merge the result with bus_routes_filtered on 'Route' to get route details
bus_routes_filtered = bus_routes[bus_routes['Mode'] == 'bus']
final_merged_data = pd.merge(merged_data, bus_routes_filtered, left_on='Route', right_on='Line Name', how='left')

# Print the merged data to verify
print(final_merged_data.head())

# Optionally, save the final merged data to a new Excel file
final_merged_data.to_excel('final_merged_data.xlsx', index=False)
```

```python
import pandas as pd

# Load the Excel file
file_path = 'all_merged_buses_data.xlsx'
df = pd.read_excel(file_path)

# Create a new DataFrame to store the separated entries
separated_entries = []

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    # Ensure the 'Routes ' column is a string and handle missing values
    routes_str = str(row['Routes ']) if pd.notna(row['Routes ']) else ''
    # Split the routes by comma
    routes = routes_str.split(', ')
    # Create a new row for each route
    for route in routes:
        new_row = row.copy()
        new_row['Routes '] = route
        separated_entries.append(new_row)

# Convert the list of separated entries into a new DataFrame
separated_df = pd.DataFrame(separated_entries)

# Save the new DataFrame to a new Excel file
new_file_path = 'separated_all_merged_buses_data.xlsx'
separated_df.to_excel(new_file_path, index=False)

print(f"Data has been processed and saved to {new_file_path}")
```
